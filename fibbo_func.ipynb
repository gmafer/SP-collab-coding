```python
def fibbonaci(n):
    """
    Arguments: 
    `n`: the nth element of the fibonacci series
    
    Output:
    `element`: the fibonnaci number corresponding to the nth element of the series
    """
    # Your code here

    
    if n<= 1:
        return(n)
    else:
        return(fibbonaci(n-1) + fibbonaci(n-2))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('parkinsons.data', 
                 dtype = { # indicate categorical variables
                     'status': 'category'})
df.head(5)
```

```python
import pandas as pd

#remove all "phon_R01"
df['new_name'] = df['name'].str.replace('phon_R01_', '')

# split 'new_name' into 'subject' and 'trial'
df[['subject', 'trial']] = df['new_name'].str.split('_', 1, expand=True)

# place new columns at the beginning
df = df[['subject', 'trial'] + [col for col in df.columns if col not in ['subject', 'trial']]]

# drop the 'new_name' column
df.drop('new_name', axis=1, inplace=True)
df.drop('name', axis=1, inplace=True)

df.head(5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def scat_plt(var1, var2, groups):

    # check if 'subject' is a categorical type, convert it if not
    if not pd.api.types.is_categorical_dtype(df['subject']):
        df['subject'] = pd.Categorical(df['subject'])

    selected_data = df[df['subject'].isin(groups)]

    colors = plt.cm.get_cmap('tab10', len(groups))
    
    for i, subject in enumerate(groups):
        subject_data = selected_data[selected_data['subject'] == subject]
        plt.scatter(subject_data[var1], subject_data[var2], label=f'Subject {subject}', color=colors(i))

    
    plt.xlabel(var1)
    plt.ylabel(var2)
    
    plt.legend()

    plt.show()
```

```python
scat_plt('MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', ['S01','S02'])
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def normalize(df, op):

    norm_df = df.copy()

    # select numeric columns
    num_cols = df.select_dtypes(include=['number']).columns

    # exclude categorical variables 
    non_categ_cols = [col for col in num_cols if not pd.api.types.is_categorical_dtype(df[col])]

    if op == 0:  # Z-score 
        scaler = StandardScaler()
        norm_df[non_categ_cols] = scaler.fit_transform(df[non_categ_cols])

    elif op == 1:  # Min-max 
        scaler = MinMaxScaler()
        norm_df[non_categ_cols] = scaler.fit_transform(df[non_categ_cols])

    return norm_df

normalized_df = normalize(df, 1)
normalized_df.head()
```
